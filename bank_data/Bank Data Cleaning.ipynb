```python
df = pd.read_csv('./bank.csv')
```

```python
columns = [col for col in df.columns]
columns
```

```python
col_list = [col.split(';') for col in df.columns ]
col_list
```

```python
flat_col_list = [element.replace("\"","") for innerList in col_list for element in innerList]
flat_col_list
```

```python
# split column into multiple columns by delimiter 
df = df['age;"job";"marital";"education";"default";"balance";"housing";"loan";"contact";"day";"month";"duration";"campaign";"pdays";"previous";"poutcome";"y"'].str.split(';', expand=True)
df.head()
```

```python
df.columns = flat_col_list
df.head()
```

```python
df.info()
```

```python
def get_unique_values(dataframe):
    cols = dataframe.columns
    for col in cols:
        print("---------------",col,"---------------")
        print(dataframe[col].unique())
        print("\n\n\n")
    
```

```python
newcols = df.columns
newcols
```

```python
for col in newcols:
    print("---------------",col,"---------------")
    print(df[col].unique())
    print("\n\n\n")
```

```python
df = df.applymap(lambda x: x.replace('"', ''))
df.head()
```

```python
get_unique_values(df)
```
