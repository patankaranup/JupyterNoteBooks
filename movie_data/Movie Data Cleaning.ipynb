```python
def get_percent_miss_value_cols(df):
    cols = df.columns
    
    for col in cols:
        percent_missing = df[col].isnull().sum() * 100 / len(df[col])
        print("------------"+col+"------------"+"\n\n"+f"Missing Percentage :{percent_missing}\n\n")
#     if sort == 0:
#         dict1 = sorted(dict1.items(),key=lambda x:x[1],reverse=False)
#         return dict1
#     else:
#         dict1= sorted(dict1.items(),key=lambda x:x[1],reverse=True)
  
```

```python
def get_unique_values(dataframe):
    cols = dataframe.columns
    for col in cols:
        print("---------------",col,"---------------")
        print(dataframe[col].unique())
        print("\n\n\n")
    
```

```python
def change_cols_to_datetime(df,cols):
    for col in cols:
        df[col]=pd.to_datetime(df[col])
```
